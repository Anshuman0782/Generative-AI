```python
import langchain
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage,SystemMessage
```

```python
model=init_chat_model("groq:llama-3.1-8b-instant")
model
```

```python
##Creating a system message
messages = [
    SystemMessage("You are a helpful AI assistant"),
    HumanMessage("What is Machine Learning")
    
]
```

```python
## invoke the model
response = model.invoke(messages)
response
```

```python
print(response.content)
```

```python
for chunk in model.stream(messages):
    print(chunk.content,end="", flush=True) 
```

Dynamic Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate

## create translation app

translation_template = ChatPromptTemplate.from_messages([
("system","You are a professional translator. Translate the follow text {text} from {source_language} to {target_language}. Maintain the tone and style of the original text."),
("user","{text}")
] )


## using the template
prompt = translation_template.invoke({
    "text":"Hello, how are you?",
   "source_language":"English",
    "target_language":"Bengali"
    }
)
```

```python
translated_response=model.invoke(prompt)
print(translated_response.content)
```
